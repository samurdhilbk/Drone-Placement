```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import matplotlib.patches as patches
```

```python
#draw with bounding rectangles
f = open("output_rects.txt", "r")
lines = f.readlines()
f.close()
fig, ax = plt.subplots()
for i, line in enumerate(lines):
    if i==0: N, r, L = list(map(float, line.strip().split(" ")))
    else:
        x, y, dx, dy = list(map(float, line.strip().split(" ")))
        rect = patches.Rectangle((x,y),dx,dy,linewidth=1,edgecolor='b',facecolor='none')
        ax.add_patch(rect)
f = open("output_circles.txt", "r")
lines = f.readlines()
f.close()
for i, line in enumerate(lines):
    if i==0: N, r, L = list(map(float, line.strip().split(" ")))
    else:
        x, y = list(map(float, line.strip().split(" ")))
        circle1 = plt.Circle((x, y), r, color='r')
        ax.add_patch(circle1)
ax.set_xlim([0, L])
ax.set_ylim([0, L])
ax.set_aspect('equal', adjustable='box')
print(N, r, L)
```
